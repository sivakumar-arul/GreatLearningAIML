# Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.

```python
def Goal_diff_count():
  grouped_team=lal.groupby(['Team']).agg('sum')
  print(grouped_team.GoalDiff)
  print ('\n\n')
  print ('Maximum Goal Diff Team : ')
  print (lal[lal.GoalDiff==lal.GoalDiff.max()]['Team'])
  print ('\n\n')
  print ('Minimum Goal Diff Team : ')
  print (lal[lal.GoalDiff==lal.GoalDiff.min()]['Team'])
    
Goal_diff_count()
```

# Create a new column with the name “Winning Percent” and append it to the data set 
# Percentage of Winning = (GamesWon / GamesPlayed)*100
# If there are any numerical error, replace it with 0%
# Print the top 5 teams which have the highest Winning percentage

```python
  filter1 = lal['GamesWon'].apply(pd.to_numeric)>0 
  filter2 = lal['GamesPlayed'].apply(pd.to_numeric)>0 
  lal.where(filter1 & filter2)
  lal['Winning Percent'] = (lal['GamesWon'].apply(pd.to_numeric)/lal['GamesPlayed'].apply(pd.to_numeric)) * 100
  lal['Winning Percent'].replace(np.nan,0,inplace=True)
  lal
```

```python
lal.nlargest(5,'Winning Percent')['Team']
```
